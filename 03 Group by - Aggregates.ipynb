## Group bys

A group by operation ( similar to group by in sql ) typically
- Splits the data
- Applies a function to the individual groups
- Combines the individual results

The function applied can be 
- aggregations like Sum, Mean etc.
- transformations like concating the strings in the group, fill nas in the group with a specific value, etc.
- Filtering to remove groups now following a condition as a group. 

Let's see all the use cases

```python
trips_data.head()
```

```python
trip_group = trips_data.groupby('start_station_id')
```

```python
type(trip_group)
```

```python
trip_group.groups
```

```python
trip_group.count()
```

```python
trip_group_2 = trips_data.groupby(['start_station_id','end_station_id'])['duration']
```

```python
type(trip_group_2)
```

```python
trip_group_2.groups
```

```python
trip_group_2.mean()
```

```python
# multiple agrregations on same column
trips_data.groupby(['start_station_id'])['duration'].agg(['mean','min','max']) 
```

```python
# multiple aggregations on different columns
trips_data.groupby(['start_station_id']).agg({'duration':['mean', 'median'], 'end_station_id':'nunique'})
```

### What happens to the index and sort order ?

```python
trips_data.groupby(['start_station_id']).agg({'duration':['mean', 'median'], 'end_station_id':'nunique'}).index
```

```python
trips_data.groupby(['start_station_id','end_station_id'])['duration'].mean()
```

> Group by objects usually have a multi level index or what is called **Hierarchical Index**

```python
trips_data.groupby(['start_station_id','end_station_id'])['duration'].mean().index
```

```python
trips_data.groupby(['start_station_id','end_station_id'])['duration'].mean().loc[2]
```

```python
trips_data.groupby(['start_station_id','end_station_id'])['duration'].mean().loc[2, 4]
```

```python
trips_data.groupby(['start_station_id','end_station_id']).mean().reset_index() # to reset the index
```

```python
trips_data_group = trips_data.groupby(['start_station_id','end_station_id']).mean().reset_index()
trips_data_group.rename(columns = {'id':'id_mean','duration':'duration_mean', 'bike_id': 'bike_id_mean'})
```

> Group by default sorts it by the grouped fields in ascending order

#### Can try with *sort = False* for possible performance improvement

```python
trips_data.groupby(['start_station_id','end_station_id'], sort= False).mean()
```

### Accessing attributes of datetime objects

Every date time object has attributes generated at the time of createing the object. These attributes can be accessed using the `.dt` operator. 

You can checkout other attributes at: http://pandas.pydata.org/pandas-docs/stable/basics.html#dt-accessor


- Common operations include extracting parts of the date object
    - hour, minutes, seconds
    - day of week (dow), week of the year

Can use library `datetime` for more operations on dates
- Add time deltas to the existing objects
- Format the dates to required formats

Lets use some of them for understanding the data better

```python
%time
trips_data['start_hour']= trips_data.start_date.dt.hour
```

```python
%time
trips_data['start_date_date']= trips_data.start_date.dt.date
```

```python
%time
trips_data['start_dow'] = trips_data.start_date.dt.dayofweek
```

```python
# Similarly, 

trips_data['end_hour']= trips_data.end_date.dt.hour
trips_data['end_date_date']= trips_data.end_date.dt.date
trips_data['end_dow'] = trips_data.end_date.dt.dayofweek
```

### Summary

#### Group by operations

- Splits-Applies-Combines the data to give required results. 
- Multiple aggregates can be performed on multiple columns in one operation
- index has to be reset else the grouped by columns become the index(in case more than 1 column, then multi index is formed)
- sort order is by default ascending on the grouped by columns

#### Date time operations 

- Various attributes of the datetime object can be accessed using `.dt` accessor object. 
- Rest of the attributes can be found here - http://pandas.pydata.org/pandas-docs/stable/basics.html#dt-accessor
